# üêº Aula 1: An√°lise de Dados com Pandas
Nesta aula, aprendemos a carregar e explorar uma base de dados de sal√°rios de profissionais da √°rea de dados (Data Science, Engenharia, etc).

## 1. Importando a Biblioteca
O **Pandas** √© a principal ferramenta do Python para manipula√ß√£o de dados. Por conven√ß√£o, importamos com o apelido `pd`.

```python
import pandas as pd

# O 'pd' √© um alias (apelido) para n√£o precisarmos digitar "pandas" toda hora
```

## 2. Carregamento dos Dados
Lemos o arquivo CSV direto do GitHub usando a fun√ß√£o `pd.read_csv`.

```python
# URL do arquivo CSV (Raw)
url = "https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv"

# Carregando para um DataFrame (df)
df = pd.read_csv(url)

# Visualizando as 5 primeiras linhas para ver se deu certo
df.head()
```

## 3. Explora√ß√£o Inicial
Precisamos entender o tamanho da base e os tipos de dados (num√©ricos ou texto).
* `info()`: Mostra colunas e tipos.
* `shape`: Mostra (linhas, colunas).

```python
# Informa√ß√µes gerais sobre tipos de dados e valores nulos
df.info()
```

```python
# Verificando o tamanho da base (Linhas, Colunas)
print(f"A base tem {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

## 4. Renomeando Colunas
Para facilitar a an√°lise, traduzimos os nomes das colunas do ingl√™s para o portugu√™s.

```python
# Dicion√°rio de tradu√ß√£o
mapa_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa_loc',
    'company_size': 'tamanho_empresa'
}

# Aplicando a renomea√ß√£o
df.rename(columns=mapa_colunas, inplace=True)

# Conferindo
df.head()
```

## 5. Estat√≠sticas Descritivas
O m√©todo `describe` nos ajuda a entender tanto os dados num√©ricos (m√©dia, m√°ximo, m√≠nimo) quanto os categ√≥ricos (quem aparece mais).

```python
# Estat√≠sticas dos dados num√©ricos (Sal√°rio, Ano)
df.describe()
```

```python
# Estat√≠sticas dos dados de texto (Categorias)
# include='object' for√ßa o pandas a analisar as strings
df.describe(include='object')
```
